#Recuperar la información del archivo

```python
import pandas as pd

# Cargar el archivo CSV
file_path = '/content/country_vaccinations.csv'
df = pd.read_csv(file_path)

# Mostrar las primeras filas del dataframe
df.head()

```

# Obtener la estructura y tipos de dato de cada columna

```python
# Mostrar la estructura y tipos de datos
df.info()

# Convertir la columna 'date' a type datetime64
df['date'] = pd.to_datetime(df['date'])

# Verificar la conversión
df.info()
```

# Cantidad de vacunas aplicadas de cada compañía

```python
# Agrupar por la columna vaccines sumando las vacunas aplicdas
vacunas_por_compania = df.groupby('vaccines')['total_vaccinations'].sum().reset_index()

# Mostrar los resultados
vacunas_por_compania
```

# Cantidad de vacunas aplicadas en todo el mundo

```python
# Sumar todas las vacunas aplicadas en todo el mundo
total_vacunas_aplicadas = df['total_vaccinations'].sum()

# Mostrar el resultado
print("Total de vacunas aplicadas en todo el mundo:",total_vacunas_aplicadas)
```

#Promedio de vacunas aplicadas por país

```python
# Calcular el promedio de vacunas aplicadas por país
promedio_vacunas_por_pais = df.groupby('country')['total_vaccinations'].mean().reset_index()

# Mostrar los resultados
print("Promedio de vacunas aplicadas por país:\n")
promedio_vacunas_por_pais
```

# Cantidad de vacunas aplicadas el día 29/01/21

```python
# Filtrar los datos swl 29/01/21
vacunas_29_enero = df[df['date'] == '2021-01-29']['total_vaccinations'].sum()

# Mostrar el resultado
print("Total de vacunas aplicadas el 29 de Enero de 2021 en todo el mundo:", vacunas_29_enero)
```

# Dataframe nuevo llamado conDiferencias

```python
# Crear una nueva columna 'diferencias'
df['diferencias'] = df['daily_vaccinations'] - df['daily_vaccinations_raw']

# Crear el nuevo dataframe de nombre conDiferencias
conDiferencias = df.copy()

# Mostrar las primeras filas del nuevo datframe
conDiferencias.head()
```

# Obtener la diferencia de tiempo

```python
# Obtener las fechas mas recientes y mas antiguas
fecha_mas_reciente = df['date'].max()
fecha_mas_antigua = df['date'].min()

# Calcular la diferencia
diferencia_tiempo = fecha_mas_reciente - fecha_mas_antigua

# Mostrar el resultado
print("periodo de tiempo entre el registro con fecha más reciente y el mas antiguo:",diferencia_tiempo)
```

# Dataframe nuevo llamado conCantidad

```python
# Crear una nueva columna canVac separando la columna vaccines
df['canVac'] = df['vaccines'].apply(lambda x: len(x.split(', ')))

# Crear el nuevo dataframe 'conCantidad'
conCantidad = df.copy()

# Mostrar las primeras filas del nuevo dataframe
conCantidad.head()
```

#Dataframe llamado antes20

```python
# Filtrar los registros antes del 20 de diciembre de 2020
antes20 = df[df['date'] < '2020-12-20']

# Mostrar las primeras filas del nuevo dataframe
antes20.head()
```

# Dataframe llamado pfizer

```python
# Filtrar los registros relacionados con la vacuna Pfizer
pfizer = df[df['vaccines'].str.contains('Pfizer')]

# Mostrar las primeras filas del nuevo dataframe
pfizer.head()
```

#Almacenar los dataframes

```python
# Guardar los dataframes en un archivo Excel
with pd.ExcelWriter('/content/resultadosReto.xlsx') as writer:
    conDiferencias.to_excel(writer, sheet_name='conDiferencias', index=False)
    conCantidad.to_excel(writer, sheet_name='conCantidad', index=False)
    antes20.to_excel(writer, sheet_name='antes20', index=False)
    pfizer.to_excel(writer, sheet_name='pfizer', index=False)
```
